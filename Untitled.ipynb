```python
import duckdb
import plotly.express as px
import ipywidgets as widgets
from IPython.display import display

con = duckdb.connect('/workspaces/github-retention-analytics/data/github.duckdb')

# Get available cohort weeks for the dropdown
weeks = con.execute("""
    SELECT DISTINCT cohort_week::VARCHAR as week 
    FROM mart_retention_cohorts 
    ORDER BY week
""").df()['week'].tolist()

# Create dropdown
week_selector = widgets.SelectMultiple(
    options=weeks,
    value=weeks,
    description='Cohort Week:',
    layout=widgets.Layout(height='150px')
)

# Create output area
out = widgets.Output()

def update_chart(change):
    with out:
        out.clear_output()
        selected = list(week_selector.value)
        
        cohorts = con.execute(f"""
            SELECT * FROM mart_retention_cohorts
            WHERE cohort_week::VARCHAR IN ({str(selected)[1:-1]})
            ORDER BY cohort_week, weeks_since_first
        """).df()
        
        fig = px.line(cohorts, 
                      x='weeks_since_first', 
                      y='retention_pct',
                      color='cohort_week',
                      title='Cohort Retention by Week',
                      labels={
                          'weeks_since_first': 'Weeks Since First Seen',
                          'retention_pct': '% Retained'
                      })
        fig.show()

week_selector.observe(update_chart, names='value')

display(week_selector, out)
update_chart(None)  # Show initial chart
```

```python
segments = con.execute("""
    SELECT engagement_segment, COUNT(*) as users
    FROM mart_engagement_segments
    GROUP BY engagement_segment
    ORDER BY users DESC
""").df()

fig2 = px.pie(segments, values='users', names='engagement_segment',
              title='User Engagement Segments', hole=0.4)
fig2.show()
```

```python
events = con.execute("""
    SELECT event_type, COUNT(*) as count
    FROM stg_events
    GROUP BY event_type
    ORDER BY count DESC
    LIMIT 10
""").df()

fig4 = px.bar(events, x='event_type', y='count',
              title='Top GitHub Event Types',
              color='count', color_continuous_scale='blues')
fig4.show()
```

```python
repos = con.execute("""
    SELECT repo_name, COUNT(*) as events
    FROM stg_events
    GROUP BY repo_name
    ORDER BY events DESC
    LIMIT 15
""").df()

fig5 = px.bar(repos, x='events', y='repo_name',
              orientation='h',
              title='Most Active Repositories',
              color='events', color_continuous_scale='greens')
fig5.show()
```
